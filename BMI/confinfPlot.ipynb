```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
hitrate = pd.read_csv('my_res.csv')
hitrate.head()
coefs = list(hitrate.columns)[2:]
hitrate = hitrate.transpose()
```

```python
len(coefs)
```

```python
sizes = hitrate.iloc[1]
plt_counter = 0
for i in range(2,hitrate.shape[0]):
    plt.plot(sizes,hitrate.iloc[i])
    plt.plot(sizes, np.full(len(hitrate.iloc[i]), np.mean(hitrate.iloc[i])))
    plt_counter += 1
    if (plt_counter == 3):
        plt_counter = 0
        ax = plt.gca()
        ax.set_ylim(0.9,1)
        plt.show()
```
