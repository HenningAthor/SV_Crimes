```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = 'r_res'
hitrate = pd.read_csv(dataset + '.csv')
hitrate.head()
coefs = list(hitrate.columns)[2:]
coefs = [ c[2:-1] if c.startswith('I(') else c for c in coefs]
hitrate = hitrate.transpose()
```

```python
colors = {0: 'red', 1: 'blue', 2 : 'green'}
sizes = hitrate.iloc[1]
plt_counter = 0
for i in range(2,hitrate.shape[0]):
    plt.plot(sizes,hitrate.iloc[i], c=colors[plt_counter], label = 'Î²' + str(i - 2))
    plt.plot(sizes, np.full(len(hitrate.iloc[i]), np.mean(hitrate.iloc[i])), '--', c=colors[plt_counter])
    plt_counter += 1
    if (plt_counter == 3):
        plt_counter = 0
        ax = plt.gca()
        ax.set_ylim(0.9,1)
        ax.legend(loc='lower right')
        plt.savefig(dataset + str(i) + ".png")
        plt.show()
```
