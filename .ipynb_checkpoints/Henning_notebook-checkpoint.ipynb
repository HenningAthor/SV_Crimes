```r
crimes <- read.csv("data/crimes.csv")
crimes
```

```r
crimes.glm <- glm(crimes~1+prbarr+density:prbarr+as.factor(region), data=crimes, family=poisson(link="log"))
crimes.glm
```

```r
summary(crimes.glm)
```

```r
plot(crimes.glm)
```

```r
fam = poisson(link="log")
m0 = glm(crimes ~ 1, data=crimes, family=fam)
m1 = glm(crimes ~ 1+prbarr, data=crimes, family=fam)
m2 = glm(crimes ~ 1+prbarr+prbpris, data=crimes, family=fam)
m3 = glm(crimes ~ 1+prbarr+prbpris+polpc, data=crimes, family=fam)
m4 = glm(crimes ~ 1+prbarr+prbpris+polpc+density, data=crimes, family=fam)
m5 = glm(crimes ~ 1+prbarr+prbpris+polpc+density+area, data=crimes, family=fam)
m6 = glm(crimes ~ 1+prbarr+prbpris+polpc+density+area+taxpc, data=crimes, family=fam)
m7 = glm(crimes ~ 1+prbarr+prbpris+polpc+density+area+taxpc+region, data=crimes, family=fam)
m8 = glm(crimes ~ 1+prbarr+prbpris+polpc+density+area+taxpc+region+pctmin, data=crimes, family=fam)
m9 = glm(crimes ~ 1+prbarr+prbpris+polpc+density+area+taxpc+region+pctmin+pctymale, data=crimes, family=fam)
m10 = glm(crimes ~ 1+prbarr+prbpris+polpc+density+area+taxpc+region+pctmin+pctymale+wcon, data=crimes, family=fam)
m11 = glm(crimes ~ 1+prbarr+prbpris+polpc+density+area+taxpc+region+pctmin+pctymale+wcon+wsta, data=crimes, family=fam)
m12 = glm(crimes ~ 1+prbarr+prbpris+polpc+density+area+taxpc+region+pctmin+pctymale+wcon+wsta+wser, data=crimes, family=fam)
m13 = glm(crimes ~ 1+prbarr+prbpris+polpc+density+area+taxpc+region+pctmin+pctymale+wcon+wsta+wser+wtrd, data=crimes, family=fam)

m14 = glm(crimes ~ 1
          +prbarr+prbarr:as.factor(region)+prbarr:polpc+prbarr:log(density)+prbarr:pctmin+prbarr:prbpris+prbarr:pctymale+prbarr:taxpc+prbarr:log(area)
          +prbpris+prbpris:as.factor(region)+prbpris:polpc+prbpris:log(density)+prbpris:taxpc+prbpris:pctymale+prbpris:pctmin+prbpris:wcon+prbpris:wser+prbpris:wsta 
          +polpc+polpc:as.factor(region)+polpc:taxpc+polpc:pctymale+polpc:pctmin
          +log(density)+log(density):as.factor(region)
          +log(area)+area:as.factor(region)
          +taxpc+taxpc:as.factor(region)+taxpc:log(density)+taxpc:pctmin+taxpc:pctymale+taxpc:wcon+taxpc:wser+taxpc:wfir
          +as.factor(region)+as.factor(region):density
          +pctmin+pctmin:as.factor(region)+pctmin:polpc+pctmin:density+pctmin:pctymale+pctmin:wcon+pctmin:wser+pctmin:wfir
          +pctymale+pctymale:as.factor(region)+pctymale:wcon+pctymale:wser+pctymale:wtrd+pctymale:wfir
          +log(wcon)+wcon:as.factor(region)+wcon:log(density)
          +log(wsta)+wsta:as.factor(region)
          +log(wser)+wser:as.factor(region)+wser:taxpc
          +log(wtrd)+log(wtrd):as.factor(region)
          +log(wfir)+wfir:as.factor(region),
          data=crimes, family=fam)

anova(m0,m1,m2,m3,m4,m5,m6,m7,m8,m9,m10,m11,m12,m13,m14, test="LRT")
# anova(m0,m14, test="LRT")

mean((crimes$crimes - predict(m14, newdata=crimes, type="response"))^2)
x = data.frame(real_crimes = crimes$crimes, pred_crimes=predict(m14, newdata=crimes, type="response"))
x$pred_crimes = round(x$pred_crimes)
x$diff = x$real_crimes-x$pred_crimes
x$abs_diff = abs(x$diff)
x$abw = x$abs_diff / x$real_crimes 
mean(x$abs_diff)
mean(x$real_crimes)
mean(x$abw)
```
