```r
crimes <- read.csv("../data/crimes.csv")
head(crimes)
```

```r
# install.packages("dplyr")
require(dplyr)

rnd_design_matrix <- function(original_data, num_rows) {
    des_mx <- data.frame(original_data) # copy the data frame
    des_mx <- des_mx[0,] # delete all rows
    
    for (i in 1 : num_rows) {
        # Add the row
        des_mx <- rbind(des_mx, sample_n(original_data, 1))
    }
    
    return(des_mx)
}

add_to_design_matrix <- function(curr_design_matrix, original_data, num_rows){
    des_mx <- data.frame(curr_design_matrix) # copy the data frame
    
    for (i in 1 : num_rows) {
        # Add the row
        des_mx <- rbind(des_mx, sample_n(original_data, 1))
    }
    
    return(des_mx)
}
```

```r
design_matrix <- rnd_design_matrix(crimes, 100)
design_matrix
```

```r
evaluate_observations <- function(observed_data){
    fam = poisson("sqrt")
    simple_glm <- glm(crimes ~ 1 + density + area + as.factor(region), data=observed_data, family=fam)
    x = data.frame(real_crimes = observed_data$crimes, pred_crimes=predict(simple_glm, newdata=observed_data, type="response"))
    
    x$pred_crimes = round(x$pred_crimes)
    x$abs_diff = abs(x$real_crimes - x$pred_crimes)
    x$rel_diff = x$abs_diff / x$real_crimes
    
    sep = sum((x$pred_crimes - x$real_crimes)^2)
    msep = sep/nrow(observed_data)
    rmsep = sqrt(msep)
    
    
    df = data.frame(real_crimes = mean(x$real_crimes), pred_crimes = mean(x$pred_crimes), abs_diff = mean(x$abs_diff), rel_diff = mean(x$rel_diff), sep=sep, msep=msep, rmsep=rmsep)
    
    return(df)
}

experiment <- function(sample_sizes, original_data){
    df <- as.data.frame(matrix(ncol = 8))
    df = df[-1,]
    colnames(df) <- c("sample_size","real_crimes","pred_crimes","abs_diff", "rel_diff", "SEP", "MSEP", "RMSEP")
    
    for (i in sample_sizes) {
        des_mx <- rnd_design_matrix(original_data, i)
        res <- evaluate_observations(des_mx)
        df <- rbind(df, data.frame(sample_size = i, real_crimes = res$real_crimes, pred_crimes = res$pred_crimes, abs_diff = res$abs_diff, rel_diff = res$rel_diff, SEP=res$sep, MSEP=res$msep, RMSEP=res$rmsep))
    }
    return(df)
}

res_exp <- experiment(seq(10, 180, by = 10), crimes)
res_exp
```
