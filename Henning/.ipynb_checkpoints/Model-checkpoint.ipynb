```r
d_cross_validation <- function(d, dataset) {
    index <- rep(1:d, length.out=nrow(dataset))
    index <- sample(index)
    
    sep_list <- rep(0, length.out=10)
    
    for (i in 1:d) {
        test_set <- dataset[index==i, ]
        train_set <- dataset[index!=i, ]
        
        m_list <- list(glm(crimes~1+prbarr, data=train_set, family=poisson(link="log")),
                       glm(crimes~1+prbarr+area, data=train_set, family=poisson(link="log")),
                       glm(crimes~1+prbarr+area, data=train_set, family=poisson(link="log")),
                       glm(crimes~1+region, data=train_set, family=poisson(link="log")),
                       glm(crimes ~ region + area + I(sqrt(density)) + I(sqrt(pctmin)) + polpc + I(polpc^2) + I(log(prbarr)) + I(sqrt(wcon)) + wtrd + I(wtrd^2), data=train_set, family=poisson(link="log")),
                       glm(crimes ~ region + area + I(sqrt(density)) + I(sqrt(pctmin)) + wtrd + I(wtrd^2), data=train_set, family=poisson(link="log")),
                       glm(crimes ~ I(density*area) + region + area + I(sqrt(density)) + I(sqrt(pctmin)) + wtrd + I(wtrd^2), data=train_set, family=poisson(link="log")),
                       glm(crimes ~ I(density*area) + region, data=train_set, family=poisson(link="log")))
        
        for (j in 1:length(m_list)){
            model <- m_list[[j]]
            pred <- predict(model, newdata=test_set, type="response")
            sep_list[j] <- sep_list[j] + sum((test_set$crimes - pred)^2)
        }
    }
    
    msep_list = sep_list / d
    rmsep_list = sqrt(msep_list)
    
    return(data.frame(SEP=sep_list, MSEP=msep_list, RMSEP=rmsep_list))
}

exp <- function(n, d, dataset){
    df <- d_cross_validation(d, dataset)
    for (i in 1:(n-1)){
        df_temp <- d_cross_validation(d, dataset)
        for (k in 1:ncol(df)){
            for (j in 1:nrow(df)){
                df[j,k] <- df[j,k] + df_temp[j,k]
            }
        }
    }
    return (df/n)
}

a = exp(100, 9, crimes)
a
```
