```r
crimes = read.csv("../data/crimes.csv")

crimes
```

```r
glm1 = glm(crimes ~ I(log(prbarr)), data=crimes, family=poisson(link="log"))
```

```r
options(repr.plot.width = 5, repr.plot.height = 5)

par(mfrow = c(2,2))

plot(crimes ~ I(log(prbarr)), data=crimes)
plot(crimes ~ I(prbarr^2), data=crimes)
plot(crimes ~ prbarr + I(prbarr^2), data=crimes)
plot(crimes ~ I(log(prbarr)), data=crimes)

options(repr.plot.width = 5, repr.plot.height = 5)
```

```r
plot(crimes~prbarr, data=crimes)
abline(lm(crimes~prbarr, data=crimes),lwd=2)
```

```r
m1 = lm(crimes~I(log(prbarr)), data=crimes)
m2 = lm(crimes~I(sqrt(prbarr)), data=crimes)
m3 = lm(crimes~prbarr+I(prbarr^2), data=crimes)
m4 = lm(crimes~prbarr+I(prbarr^2)+I(prbarr^3), data=crimes)
x <- seq(0, 0.7, 0.01)

plot(crimes~prbarr, data=crimes)
abline(lm(crimes~prbarr, data=crimes),lwd=2)

options(repr.plot.width = 10, repr.plot.height = 10)

par(mfrow = c(2,2))

plot(crimes~prbarr, data=crimes)
points(x, predict(m1, newdata = data.frame(prbarr=x)),type="l", col=2, pch=16, lwd=2) #rot

plot(crimes~prbarr, data=crimes)
points(x, predict(m2, newdata = data.frame(prbarr=x)),type="l", col=3, pch=16, lwd=2) #grn

plot(crimes~prbarr, data=crimes)
points(x, predict(m3, newdata = data.frame(prbarr=x)),type="l", col=4, pch=16, lwd=2) #dunkelblau

plot(crimes~prbarr, data=crimes)
points(x, predict(m4, newdata = data.frame(prbarr=x)),type="l", col=5, pch=16, lwd=2) #cyan

options(repr.plot.width = 5, repr.plot.height = 5)
```

```r
a = "crimes"
b = "prbarr"

m0 = glm(as.formula(paste0(a , "~", b)), data=crimes, family=poisson(link="log"))
m1 = glm(as.formula(paste0(a, "~ I(log(", b,"))")), data=crimes, family=poisson(link="log"))
m2 = glm(crimes~I(sqrt(prbarr)), data=crimes, family=poisson(link="log"))
m3 = glm(crimes~prbarr+I(prbarr^2), data=crimes, family=poisson(link="log"))
m4 = glm(crimes~prbarr+I(prbarr^2)+I(prbarr^3), data=crimes, family=poisson(link="log"))
x <- seq(0, 0.7, 0.01)

options(repr.plot.width = 4.5, repr.plot.height = 4.5)

plot(crimes~prbarr, data=crimes, main="Normal Model")
points(x,predict(m0, newdata = data.frame(prbarr=x), type="response"), type="l",lwd=3)

options(repr.plot.width = 10, repr.plot.height = 10)

par(mfrow = c(2,2))

plot(crimes~prbarr, data=crimes, main="Log Model")
points(x,predict(m1, newdata = data.frame(prbarr=x), type="response"), type="l",lwd=3,col=2) #rot

plot(crimes~prbarr, data=crimes, main="Sqrt Model")
points(x,predict(m2, newdata = data.frame(prbarr=x), type="response"), type="l",lwd=3,col=3) #grn

plot(crimes~prbarr, data=crimes, main="Square Model")
points(x,predict(m3, newdata = data.frame(prbarr=x), type="response"), type="l",lwd=3,col=4) #dunkelblau

plot(crimes~prbarr, data=crimes, main="Cubic Model")
points(x,predict(m4, newdata = data.frame(prbarr=x), type="response"), type="l",lwd=3,col=5) #cyan

options(repr.plot.width = 5, repr.plot.height = 5)
```
