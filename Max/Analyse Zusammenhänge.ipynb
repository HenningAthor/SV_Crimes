```r
set.seed(1)
index = rep(1:10, length.out=90)
index = sample(index)

SEP0 = SEP1 = SEP2 = SEP3 = SEP4 = SEP5 = SEP6 = SEP7 = SEP8 = 0

for (i in 1:10)
{
    ### Zerlegung
    crimes.test = crimes[index==i,]
    crimes.train = crimes[index!=i,]
    
    #print(crimes.test)
    #print(crimes.train)
    
    ### Parameterschtzung
    crimesglm0 = glm(crimes ~ region, data=crimes.train, family=poisson(link="log"))
    crimesglm1 = glm(crimes ~ region + area + density + pctmin + polpc + prbarr + prbpris + wcon + wfir + wser + wtrd, data=crimes.train, family=poisson(link="log"))
    crimesglm2 = glm(crimes ~ region + area + density + pctmin + prbarr + wcon + wfir + wser + wtrd, data=crimes.train, family=poisson(link="log"))
    crimesglm3 = glm(crimes ~ region + area + I(sqrt(density)) + I(sqrt(pctmin)) + polpc + I(polpc^2) + I(log(prbarr)) + prbpris + I(prbpris^2) + I(sqrt(wcon)) + I(log(wfir)) + I(sqrt(wser)) + wtrd + I(wtrd^2), data=crimes.train, family=poisson(link="log"))
    crimesglm4 = glm(crimes ~ region + area + I(sqrt(density)) + I(sqrt(pctmin)) + polpc + I(polpc^2) + I(log(prbarr)) + I(sqrt(wcon)) + wtrd + I(wtrd^2), data=crimes.train, family=poisson(link="log"))
    #aus regsubset mit crossvalidation mit "crimes ~ ."
    crimesglm5 = glm(crimes ~ region * density + area, data=crimes.train, family=poisson(link="log"))
    # aus regsubset mit crossvalidation mit "crimes ~ .^2"
    crimesglm6 = glm(crimes ~ density:wcon + area:wcon, data=crimes.train, family=poisson(link="log"))
    # Model 4 ber regsubset mit crossvaldiation verbessern
    crimesglm4New = glm(crimes ~ region + area + I(sqrt(density)) + I(sqrt(pctmin)) + wtrd + I(wtrd^2), data=crimes.train, family=poisson(link="log"))
    # neues Model 4 ber regsubset mit crossvaldiation verbessern und mit Interaktion -> scheint fr Interaktionen nicht sinnvoll zu sein
    crimesglm4NewInteraction = glm(crimes ~ region + area + I(sqrt(density)) + I(sqrt(pctmin)) + wtrd + I(wtrd^2) + region:I(sqrt(density)) + region:wtrd + area:I(sqrt(density)) + I(sqrt(density)):wtrd + I(sqrt(pctmin)):wtrd + wtrd:I(wtrd^2), data=crimes.train, family=poisson(link="log"))
    #crimesglm4NewInteraction = glm(crimes ~ I(sqrt(pctmin)) + I(wtrd^2) + region:I(sqrt(density)) + region:wtrd + area:I(sqrt(density)) + I(sqrt(density)):wtrd + I(sqrt(pctmin)):wtrd + wtrd:I(wtrd^2), data=crimes.train, family=poisson(link="log"))
    
    ### Prognosefehler
    SEP0 = SEP0 + sum( ( crimes.test$crimes - predict(crimesglm0, newdata=crimes.test, type="response") )^2 )
    SEP1 = SEP1 + sum( ( crimes.test$crimes - predict(crimesglm1, newdata=crimes.test, type="response") )^2 )
    SEP2 = SEP2 + sum( ( crimes.test$crimes - predict(crimesglm2, newdata=crimes.test, type="response") )^2 )
    SEP3 = SEP3 + sum( ( crimes.test$crimes - predict(crimesglm3, newdata=crimes.test, type="response") )^2 )
    SEP4 = SEP4 + sum( ( crimes.test$crimes - predict(crimesglm4, newdata=crimes.test, type="response") )^2 )
    SEP5 = SEP5 + sum( ( crimes.test$crimes - predict(crimesglm5, newdata=crimes.test, type="response") )^2 )
    SEP6 = SEP6 + sum( ( crimes.test$crimes - predict(crimesglm6, newdata=crimes.test, type="response") )^2 )
    SEP7 = SEP7 + sum( ( crimes.test$crimes - predict(crimesglm4New, newdata=crimes.test, type="response") )^2 )
    SEP8 = SEP8 + sum( ( crimes.test$crimes - predict(crimesglm4NewInteraction, newdata=crimes.test, type="response") )^2 )
}

cat( "Modell 0: ", SEP0, "\n")
cat( "Modell 1: ", SEP1, "\n")
cat( "Modell 2: ", SEP2, "\n")
cat( "Modell 3: ", SEP3, "\n")
cat( "Modell 4: ", SEP4, "\n")
cat( "Modell 5: ", SEP5, "\n")
cat( "Modell 6: ", SEP6, "\n")
cat( "Modell 7: ", SEP7, "\n")
cat( "Modell 8: ", SEP8, "\n")
```

```r
# Set seed for reproducibility
set.seed(123)
# Set up repeated k-fold cross-validation
train.control <- trainControl(method = "cv", number = 10)
# Train the model
step.model <- train(crimes ~ (region + area + I(sqrt(density)) + I(sqrt(pctmin)) + wtrd + I(wtrd^2))^2, data = crimesWithoutCounty,
                    method = "leapBackward", 
                    tuneGrid = data.frame(nvmax = 1:20),
                    trControl = train.control
                    )
step.model$results
```
