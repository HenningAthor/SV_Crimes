```r
crimes.data = read.csv('crimes.csv')
crimes.data$population = crimes.data$area * crimes.data$density
crimes.data$crime_rate = crimes.data$crime / crimes.data$population
crimes.data
```

```r
x = lm(paste('crimes~','region'), crimes.data)
#plot(paste0('crimes~','region'), crimes.data)
#plot(crimes~region,crimes.data)
plot(x,which=1)
```
